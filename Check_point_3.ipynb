# **Outliers**

Outliers adalah nilai yang jauh berbeda dari nilai lainnya dalam dataset. Nilai Outlier bisa jauh lebih rendah atau lebih tinggi. Outlier bisa terjadi karena berbagai alasan seperti faktor kesalahan maupun kejadian lain yang tidak terduga.

Melakukan pengecekan outliers

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("/content/gender_check_point_2.xlsx")
df.head()
```

# **BAR CHART**

```python

df_mean = df.groupby("gender")[["forehead_width_cm", "forehead_height_cm", "nose_wide", "nose_long", "lips_thin"]].mean()
df_mean.plot(kind="bar", figsize=(8, 5), color=['skyblue', 'salmon', 'lightgreen', 'orange', 'purple'])
plt.title("Rata-rata Karakteristik Wajah per Gender")
plt.xlabel("Gender")
plt.ylabel("Rata-rata Nilai")
plt.xticks(rotation=0)
plt.legend(title="Fitur Wajah")
plt.show()
```

# **PIE CHART**

```python
df["gender"].value_counts().plot(kind="pie", autopct="%1.1f%%", colors=['lightblue', 'lightcoral'], figsize=(6, 6))
plt.title("Distribusi Gender dalam Dataset")
plt.ylabel("")
plt.show()
```

# **SCATTER PLOT**

```python
colors = {"Male": "blue", "Female": "red"}
plt.figure(figsize=(8,5))
for gender in df["gender"].unique():
    subset = df[df["gender"] == gender]
    plt.scatter(subset["forehead_width_cm"], subset["forehead_height_cm"],
                color=colors[gender], label=gender, alpha=0.6)
plt.xlabel("Forehead Width (cm)")
plt.ylabel("Forehead Height (cm)")
plt.title("Scatter Plot Forehead Width vs Forehead Height")
plt.legend()
plt.show()

```

Bar Chart : Perbandingan Rata-rata Rasio Forehead per Gender

Insight:
Rata-rata forehead width pada laki-laki lebih besar dibandingkan perempuan, sementara forehead height lebih bervariasi pada kedua gender.
Perbedaan ini menunjukkan bahwa ukuran forehead bisa menjadi salah satu faktor dalam membedakan gender.

Tindakan:
